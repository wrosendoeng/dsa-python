```python
class Funcionarios:
    def __init__(self,nome,salario):
        self.nome = nome
        self.salario = salario
        
    def listFunc(self):
        print("O nome do funcionário'e " + self.nome + " e o salario dele é de R$ " + str(self.salario) + ",00")
```

```python
func1 = Funcionarios("Jorge",10000)
```

```python
func1.listFunc()
```
