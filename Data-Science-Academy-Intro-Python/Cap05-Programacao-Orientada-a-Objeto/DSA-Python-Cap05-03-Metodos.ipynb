```python
class Circulo:
    
    # O valor de pi é constante
    pi = 3.14
    
    def __init__(self, raio = 5):
        self.raio = raio
    
    def area(self):
        return (self.raio * self.raio) * Circulo.pi
    
    def setRaio(self, novo_raio):
        self.raio = novo_raio
    
    def getRaio(self):
        return self.raio
```

```python
circ = Circulo()
```

```python
circ.getRaio()
```

```python
circ.area()
```

```python
circ.setRaio(3)
```

```python
circ.getRaio()
```

```python
circ.area()
```

```python
print("o novo raio é: ", circ.getRaio())
```
