```python
class Livro():
    
    def __init__(self,titulo,isbn):
        self.titulo = titulo
        self.isbn = isbn
        print('Construtor chamado para criar um objeto desta classe')
        
    def imprime(self):
        print("Foi criado o livro %s e ISBN %d" %(self.titulo, self.isbn))
```

```python
Livro1 = Livro('O Monge e o Executivo',99888888)
```

```python
type(Livro1)
```

```python
Livro1.titulo
```

```python
Livro1.imprime()
```

```python
Livro2 = Livro("A Menina que roubava Livros",77666666)
```

```python
Livro2.imprime()
```
