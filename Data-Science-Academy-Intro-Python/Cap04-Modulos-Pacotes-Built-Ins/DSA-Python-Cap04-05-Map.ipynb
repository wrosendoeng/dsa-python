```python
def fahrenheit(T):
    return ((float(9)/5)*T+32)

def celsius(T):
    return (float(5)/9)*(T-32)
```

```python
temperaturas = [0,22,34,100]
```

```python
list(map(fahrenheit,temperaturas))
```

```python
list(map(lambda x: (5.0/9)*(x-32), temperaturas))
```
