```python
def verificaPar(num):
    if num%2 == 0:
        return True
    else:
        return False
```

```python
verificaPar(35)
```

```python
lista = list(range(0,19))
```

```python
lista
```

```python
filter(verificaPar,lista)
```

```python
list(filter(verificaPar,lista))
```

```python
list(filter(lambda x: x%2==0,lista))
```

```python
list(filter(lambda x: x > 8,lista))
```
