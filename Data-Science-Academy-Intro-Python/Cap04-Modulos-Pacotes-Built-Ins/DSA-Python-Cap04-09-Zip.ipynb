```python
d1 = {'a':1,'b':2}
d2 = {'c':3,'d':4}
```

```python
list(zip(d1,d2))
```

```python
list(zip(d1,d2.values()))
```

```python
def trocaValores(d1,d2):
    dicTemp = {}
    
    for d1key, d2val in zip(d1,d2.values()):
        dicTemp[d1key] = d2val
        
    return dicTemp
```

```python
trocaValores(d1,d2)
```
