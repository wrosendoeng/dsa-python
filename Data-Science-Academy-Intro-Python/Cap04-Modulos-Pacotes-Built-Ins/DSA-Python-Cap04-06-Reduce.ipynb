```python
from functools import reduce
```

```python
lista = [47,11,42,13]
```

```python
def soma(a,b):
    x = a + b
    return x
```

```python
reduce(soma,lista)
```

```python
lst = lista
```

```python
reduce(lambda x,y:x+y,lst)
```

```python
max_find2 = lambda a,b: a if (a > b) else b
```

```python
type(max_find2)
```

```python
reduce(max_find2,lst)
```
