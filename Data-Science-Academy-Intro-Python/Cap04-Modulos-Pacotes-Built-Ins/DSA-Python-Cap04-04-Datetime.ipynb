```python
t = datetime.time(7,43,28)
```

```python
print (t)
```

```python
print ('Hora :',t.hour)
print ('Minuto: ',t.minute)
print ('Segund: ', t.second)
print ('Microsegundo: ', t.microsecond)
```

```python
hoje = datetime.date.today()
```

```python
print(hoje)
print('data: ',hoje.ctime())
print('ano: ',hoje.year)
print('mês: ',hoje.month)
print('dia: ',hoje.day)
```

```python
d1 = datetime.date(2015,4,28)
print('d1: ',d1)
d2 = datetime.date(2021,8,8)
print('d2: ',d2)
print(d2 - d1)
```
