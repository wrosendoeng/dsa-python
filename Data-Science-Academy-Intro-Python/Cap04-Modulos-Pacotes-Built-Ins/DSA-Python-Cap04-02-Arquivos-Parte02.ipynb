# Manipulando dados CSV

```python
import csv
```

```python
with open('arquivos/numeros.csv', 'w') as arquivo:
        writer = csv.writer(arquivo)
        writer.writerow(('primeira','segunda','terceira'))
        writer.writerow((55,93,76))
        writer.writerow((62,14,86))
```

```python
with open('arquivos/numeros.csv','r') as arquivo:
    leitor = csv.reader(arquivo)
    for x in leitor:
        print ('Número de colunas:', len(x))
        print(x)
```

```python
with open('arquivos/numeros.csv','r') as arquivo:
    leitor = csv.reader(arquivo)
    dados = list(leitor)
    
print(dados)
```

```python
for linha in dados[1:]:
    print (linha)
```

# Manipulando Arquivos JSON (JavaScript Object Notation)

JSON é uma maneira de armazenar informações de forma organizada e de fácil acesso. Em poucas palavras, ele nos dá uma coleção legível de dados que podem ser acessados de forma muito lógica. Pode ser uma fonte de Big Data.

```python
# criando um dicionário

dict = {'nome': 'Guido van Rossum',
        'linguagem': 'Python',
        'similar': ['c','Modula-3','lisp'],
        'users': 1000000
       }
```

```python
for k, v in dict.items():
    print(k,v)
```

```python
import json
```

```python
json.dumps(dict)
```

```python
with open('arquivos/dados.json','w') as arquivo:
    arquivo.write(json.dumps(dict))
```

```python
with open('arquivos/dados.json','r') as arquivo:
    texto = arquivo.read()
    data = json.loads(texto)
```

```python
print (data)
```

```python
print (data['nome'])
```

```python
from urllib.request import urlopen

response = urlopen("http://vimeo.com/api/v2/video/57733101.json").read().decode('utf8')
data = json.loads(response)[0]
```

```python
print('Titulo: ',data['title'])
print("URL: ",data['url'])
print('Duracao: ',data['duration'])
print('Numero de visualizacoes: ',data['stats_number_of_plays'])
```

```python
import os

arquivo_fonte = 'arquivos/dados.json'
arquivo_destino = 'arquivos/json_data.txt'
```

```python
with open(arquivo_fonte,'r') as infile:
    text = infile.read()
    with open(arquivo_destino,'w') as outfile:
        outfile.write(text)
```

```python
open(arquivo_destino,'w').write(open(arquivo_fonte,'r').read())
```
